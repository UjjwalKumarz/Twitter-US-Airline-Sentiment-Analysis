```python
documents = []
for i in range(len(x_training)):
    text = ''
    for col in ['airline','text', 'tweet_location','user_timezone']:
        text += str(x_training[col][i])+' '
    documents.append([w for w in word_tokenize(text) if w.lower() not in stops])
documents[0:5]
```

```python
test_documents = []
for i in range(len(testing)):
    text = ''
    for col in ['airline','text', 'tweet_location','user_timezone']:
        text += str(testing[col][i])+' '
    test_documents.append([w for w in word_tokenize(text) if w.lower() not in stops])
test_documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag): #creating simple tags to pass into the lemmatizer
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        pos = pos_tag([w])                             
        clean_word = lem.lemmatize(w, get_simple_pos(pos[0][1]))
        output_words.append(clean_word.lower())
    return output_words
```

```python
import nltk
nltk.download('wordnet')
```
